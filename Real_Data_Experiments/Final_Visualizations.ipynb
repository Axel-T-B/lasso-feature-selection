```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore', category=FutureWarning)

dataset_name = "TCGA_GBM_LGG"

# 1. Load the Top 10 final table
df = pd.read_csv(f"Top_10_Genes_{dataset_name}.csv")

# 2. Display the exact table for the research paper
print("Table: Top 10 Selected Genes\n")
print(df.to_string(index=False))

# 3. Create a Bar Chart for the Presentation Slides
plt.figure(figsize=(10, 6))

# Sort by actual coefficient value so the biological direction (positive/negative) is clear
df_plot = df.sort_values(by='Coefficient', ascending=True)

# Red for negative survival correlation, Green for positive
colors = ['#d9534f' if c < 0 else '#5cb85c' for c in df_plot['Coefficient']]

sns.barplot(
    data=df_plot, 
    x='Coefficient', 
    y='Gene', 
    palette=colors
)

plt.title(f"Top 10 Discovered Biomarkers ({dataset_name})", fontsize=16, fontweight='bold')
plt.xlabel("Feature Importance Score (Coefficient)", fontsize=12)
plt.ylabel("Gene", fontsize=12)
plt.axvline(x=0, color='black', linewidth=1)
plt.tight_layout()

# Save it so you can drop it right into PowerPoint
plt.savefig("Top_10_Genes_Chart.png", dpi=300)
plt.show()
```
